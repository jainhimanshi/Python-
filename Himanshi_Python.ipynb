```python
#Map:
f_list=[2,3,4]
print(map(lambda x:x**2,f_list))
print(list(map(lambda x:x**2,f_list)))

#Using the Map function, create a list 'cube', which consists of the cube of numbers in input_list.
#if the input list is [5,6,4,8,9], the output should be [125, 216, 64, 512, 729].
input_list=[5,6,4,8,9]
print(list(map(lambda x:x**3,input_list)))

#Using the function Map, count the number of words that start with ‘S’ in input_list.
input_list2 = ["Sa","Si","Ho"]
print(list(map(lambda x:x[0]=='S',input_list2)))

#example using two input args
list1=[1,2]
list2=[3,4]
print(list(map(lambda x,y:x+y,list1,list2)))

#Filter:
#Filter' is a similar operation, but it requires the function to look for a condition and then, only returns
#those elements from the collection that satisfy the condition.
list3=[1,2,3,6]
print(list(filter(lambda x:x%3==0,list3)))

#Create a list ‘name’ consisting of the combination of the first name and the second name from list 1 and 2 respectively. 
list4=['Himanshi','mithil']
list5=['jain','verma']
name=list(map(lambda x,y:x[0:]+' '+y[0:],list4,list5))
print(name)

#Reduce 
from functools import reduce
print(reduce(lambda x,y:x+y,range(1,4))) #0,1,2,3

list5=[1,2,3,4,5]
print(reduce(lambda x,y:x if x<y else y,list5))

#Extract a list of numbers that are multiples of 5 from a list of integers named input_list.
input_list1=[5,2,10,1]
print(list(filter(lambda x:x%5==0,input_list1)))

#You are given a list of strings such as input_list =  ['hdjk', 'salsap', 'sherpa'].
#Extract a list of names that start with an ‘s’ and end with a ‘p’ (both 's' and 'p' are lowercase) in input_list.
input_list =['hdjk','salsap','sherpa']
print(list(filter(lambda x:x[0]=='s'and x[-1]=='p',input_list)))

#Using the Reduce function, concatenate a list of words in input_list, and print the output as a string.
#If input_list = ['I','Love','Python'], the output should be the string 'I Love Python'.
input_list4 = ['I','Love','Python']
print(reduce(lambda x,y:x+" "+y,input_list4))
```
